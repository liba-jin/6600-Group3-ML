```python
weather = df[["DATE", "AW1"]].copy()

# Convert DATE to datetime (important for downstream)
weather["DATE"] = pd.to_datetime(weather["DATE"])

# Extract date + hour
weather["Weather_Date"] = weather["DATE"].dt.date
weather["Weather_Hour"] = weather["DATE"].dt.hour


# Final minimal dataset
weather = weather[["Weather_Date", "Weather_Hour", "AW1"]]

weather.head()
```

```python
weather["AW1"] = weather["AW1"].astype(str).str.split(",").str[0]

weather["AW1"] = pd.to_numeric(weather["AW1"], errors="coerce")

weather = weather.dropna(subset=["AW1"])

weather.head()
```

```python
weather = weather.merge(
    weather_mapping,
    left_on="AW1",
    right_on="code",
    how="left"
)
```

```python
weather.head(10)
```

```python
# rename
weather = weather.rename(columns={
    "Weather_Date": "weather_date",
    "Weather_Hour": "weather_hour"
})

# flags + intensities
weather_flags = ["rain", "thunderstorm", "freezing_rain", "snow", "fog"]

for flag in weather_flags:
    weather[f"{flag}_flag"] = (weather["grouped_weather_event"] == flag).astype(int)
    weather[f"{flag}_intensity"] = weather["intensity"].where(
        weather["grouped_weather_event"] == flag
    )

# keep only desired columns
final_columns = ["weather_date", "weather_hour", "AW1"]
for flag in weather_flags:
    final_columns += [f"{flag}_flag", f"{flag}_intensity"]

weather = weather[final_columns]
```

```python
weather.head(10)
```

```python
#next we need to collapse this data into one row per hour. we will take the max intensity of each event type
#convert intensity to numeric severity

intensity_map = {
    "light": 1,
    "moderate": 2,
    "heavy": 3
}

weather_flags = ["rain", "thunderstorm", "freezing_rain", "snow", "fog"]

for flag in weather_flags:
    weather[f"{flag}_intensity_num"] = weather[f"{flag}_intensity"].map(intensity_map)
```

```python
#aggregate

agg_dict = {}

# Flags → max
for flag in weather_flags:
    agg_dict[f"{flag}_flag"] = "max"
    agg_dict[f"{flag}_intensity_num"] = "max"

hourly_weather = (
    weather
    .groupby(["weather_date", "weather_hour"])
    .agg(agg_dict)
    .reset_index()
)
```

```python
#convert back to labels

reverse_intensity_map = {
    1: "light",
    2: "moderate",
    3: "heavy"
}

for flag in weather_flags:
    hourly_weather[f"{flag}_intensity"] = hourly_weather[f"{flag}_intensity_num"].map(reverse_intensity_map)
```

```python
#drop numeric columns

hourly_weather = hourly_weather.drop(
    columns=[f"{flag}_intensity_num" for flag in weather_flags]
)
```
